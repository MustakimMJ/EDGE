```python
#Problem 6 (Reverse list)
a=[36, 69, 24, 32, 45, 62, 45, 93]
n=len(a)
b=n
n/=2
i=0
while(i<=n):
  a[i],a[b-i-1]=a[b-i-1],a[i]
  i+=1
print(a)
```

```python
#Problem 8(Find the maximum in a list)
a=[36, 69, 24, 32, 45, 62, 45, 93]
max_=a[0]
i=0
while i<len(a):
  if max_<a[i]:
    max_=a[i]
  i+=1
print(max_)
```
