```python
from random import randint
import time
listlen=1000000
listOfNumbers=listlen*[0]
dictOfNumbers=dict()
start_time=time.time()

for i in range(listlen):
  n=randint(1000000,9999999)
  listOfNumbers[i]=n
end_time=time.time()
time_taken=end_time-start_time
print(time_taken)
for i in range(listlen):
  n=randint(1000000,9999999)
  dictOfNumbers[n]=i
end_time=time.time()
time_taken=end_time-start_time
print(time_taken)
```
