```python
def pcheck(a):
  b=a[::-1]
  if a==b:
    print('Pallindrome')
  else:
    print('Not Pallindrome')
a=input('Enter to check Pallindrome or not: ')
pcheck(a)  #[:: means= start: stop: step]
```
