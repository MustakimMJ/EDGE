```python
#10
balance=0
while True:
  co=int(input('Enter choice:\n 1. Check Balance \n 2. Deposit Money \n 3. Withdraw money \n 4. Exit\n'))
  if co==4:
    break
  elif co==1:
    print(balance)
  elif co==2:
    a=int(input('Enter a the deposited amount: '))
    balance+=a
    print(f"Money depositt successfull \n The total balance is: {balance}")
  elif co==3:
    h=int(input('Enter the number of money you wat to withdraw: '))
    if h>balance:
      print('Insufficient Amount')
    else:
      balance-=h
      print(f"Money withdraw successfull \n The remaining balance is: {balance}")
  else:
    print('wrong choice! Try again')
```
