```python
li=[23,45,67,66,92,34,69]
def isOdd(n):
  if n%2==1:
    return n
odd_numbers=list(map(isOdd,li))
print(odd_numbers)
```

```python
nl=[23,45,67,66,92,34,69]
odd_numbers=list(filter(lambda n:n%2==1,nl))
print(odd_numbers)
```
